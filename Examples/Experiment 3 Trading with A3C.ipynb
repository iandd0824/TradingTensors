```python
from tradingtensors.Environments.CustomEnvs import OandaEnv
from tradingtensors.Agents.ActorCritic import A3CAgent


PATH= r'train\A3C\20180826-EUR_USD_H1'

#Create a variable to store a function to initialise env (use lambda)
create_env = lambda: OandaEnv(
        INSTRUMENT = "EUR_USD", 
        TIMEFRAME = "H1",
        additional_pairs = [],
        indicators = [],
        lookback_period= 3,
        trade_duration=3,
        planet_data={"Moon" : ["right asc.", "declination", "distance"]}, # Require these coordinates of Moon
        PLANET_FORWARD_PERIOD=2 #Look ahead 2 periods of Moon Coordinates 
    )


#Define agent
agent = A3CAgent(create_env, PATH)
```

```python
agent.trainGlobalNet(REWARD_THRESHOLD=10000, MAX_EPS=100)
```

```python
agent.performs(TRAIN=True)
```

```python
agent.performs(TRAIN=False)
#Still overfitting, adjust coefficient of regularization (L2)
```
